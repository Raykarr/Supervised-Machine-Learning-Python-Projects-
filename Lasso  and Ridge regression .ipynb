<font color='red'>Lasso</font> 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy import arange
from sklearn.linear_model import LassoCV
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import RepeatedKFold
```

```python
url = "https://raw.githubusercontent.com/Statology/Python-Guides/main/mtcars.csv"
df = pd.read_csv(url,index_col=0)
df
```

```python
data =df [["mpg","wt","drat","qsec","hp"]]
```

```python
x = data[["mpg","wt","drat","qsec"]]
y = data[["hp"]]
```

```python
x
```

```python
y
```

```python
cv = RepeatedKFold(n_splits=10, n_repeats=3,random_state=1)
```

```python
model = LassoCV(alphas=arange(0,1,0.01),cv=cv)
model.fit(x,y)
```

```python
print(model.alpha_)
```

#calcute the print root mean square (MSE)

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(x,y) 
rase = np.sqrt(mse) 
print("MSE value: {:.4f}".format(rase))

```python
new = [24,2.5,3.5,18.5]
```

```python
model.predict([new])
```

```python
model1=RidgeCV(alphas=arange(0,1,0.01),cv=cv)
model1.fit(x,y)
```

```python
print(model1.alpha_)
```
